```python
# Call in ls2 Data
dataframe = pd.read_csv('/nas/cee-water/cjgleason/ellie/SNiP/RFmodel/CONUS/LS2/LS2NoMatch_2000-2023.csv')
ls2 = dataframe[(dataframe.red > 0) & (dataframe.pixelCount > 5) & (dataframe.nir > 0) & (dataframe.blue > 0) & (dataframe.green > 0) & 
                      (dataframe.swir1 > 0) & (dataframe.swir2 > 0) & (dataframe.nir > 0) & (dataframe.pixelCount > 10)].drop_duplicates(subset = ['SiteID', 'date'])
ls2['date'] = pd.to_datetime(ls2['date'])
ls2 = ls2.loc[(ls2['date'] >= '2013-03-18')]

# use the map function to create a new column based on the dictionary
ls2['lat'] = ls2['SiteID'].map(my_dict)

ls2['long'] = ls2['SiteID'].map(my_dict_long)
#fusionPred = fusionPred.dropna(subset = ['huc'])


# Identify rows with values in the 'name' column that are not in the dictionary keys
#mask = ~fusionPred['SiteID'].isin(my_dict)

# Drop the rows that have values in the 'name' column not in the dictionary keys
#fusionPred.drop(fusionPredB[mask].index, inplace=True)

ls2['lat']= ls2.groupby('SiteID')['lat'].fillna(method='ffill')  # use 'bfill' for backward filling
ls2['lat']= ls2.groupby('SiteID')['lat'].fillna(method='bfill')  # use 'bfill' for backward filling

ls2['long']= ls2.groupby('SiteID')['long'].fillna(method='ffill')  # use 'bfill' for backward filling
ls2['long']= ls2.groupby('SiteID')['lat'].fillna(method='bfill')  # use 'bfill' for backward filling

ls2['Matchup'] = 'LS2_NoMatchup'
print(ls2.head())
```

```python
allMeas = pd.concat([fusionPred, allMatchups, ls2]) #, df_huc, df_SiteID
allMeas = allMeas[['SiteID', 'lat', 'long', 'Matchup']]
```

```python

# Group by two categories and count the size of each group
groupedMatchups = allMeas.groupby(['SiteID', 'Matchup'])['lat'].count().unstack().fillna(0).reset_index()

# Select the unique category1 so it only shows once but preserves the rest of the dataframe
alldf = groupedMatchups
alldf.loc[:, 'AquaLS2_Matchup'] = alldf.loc[:, 'AquaSat'] + alldf.loc[:, 'LS2']
alldf.loc[:, 'Matchup_Total'] = alldf.loc[:, 'AquaSat'] + alldf.loc[:, 'LS2'] + alldf.loc[:, 'Fusion']
alldf.loc[:, 'NoMatchup_Total'] = alldf.loc[:, 'LS2_NoMatchup'] + alldf.loc[:, 'FusionNoMatchup']
alldf.loc[:, 'AquaLS2_Total'] = alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup']
alldf.loc[:, 'Fusion_Total'] = alldf.loc[:, 'FusionNoMatchup'] + alldf.loc[:, 'Fusion']

alldf.loc[:, 'LS2GainOverAqua_Matchup'] = alldf.loc[:, 'LS2'] - alldf.loc[:, 'AquaSat']
alldf.loc[:, 'FusionGainOverAqua_Matchup'] = alldf.loc[:, 'Fusion'] - alldf.loc[:, 'AquaSat']
alldf.loc[:, 'FusionGainOverLS2_Matchup'] = alldf.loc[:, 'Fusion'] - alldf.loc[:, 'LS2']
alldf.loc[:, 'FusionGainOverAquaLS2_Matchup'] = alldf.loc[:, 'Fusion'] - alldf.loc[:, 'AquaLS2_Matchup']

alldf.loc[:, 'FusionGainOverLS2_NoMatchup'] = alldf.loc[:, 'FusionNoMatchup'] - alldf.loc[:, 'LS2_NoMatchup']
alldf.loc[:, 'FusionGainOverAquaLS2_Total'] = (alldf.loc[:, 'FusionNoMatchup'] + alldf.loc[:, 'Fusion']) - (alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup'])

alldf.loc[:, 'percentFusionAquaLS2_Matchup'] = (alldf.loc[:, 'Fusion'] / (alldf.loc[:, 'AquaLS2_Matchup'])) * 100
alldf.loc[:, 'percentFusionAqua_Matchup'] = (alldf.loc[:, 'Fusion'] / alldf.loc[:, 'AquaSat']) * 100
alldf.loc[:, 'percentFusionLS2_Matchup'] = (alldf.loc[:, 'Fusion'] / alldf.loc[:, 'LS2']) * 100

alldf.loc[:, 'percentFusionAquaLS2_NoMatchup'] = (alldf.loc[:, 'FusionNoMatchup'] / (alldf.loc[:, 'LS2_NoMatchup'])) * 100
#alldf.loc[:, 'percentFusionAquaLS2_Total'] = (((alldf.loc[:, 'FusionNoMatchup'] + alldf.loc[:, 'Fusion']) - (alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup']))
#                                              / (alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup'])) * 100
alldf.loc[:, 'percentFusionAquaLS2_Total'] = (((alldf.loc[:, 'Matchup_Total'] + alldf.loc[:, 'NoMatchup_Total']) - (alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup']))
                                              / abs(alldf.loc[:, 'AquaLS2_Matchup'] + alldf.loc[:, 'LS2_NoMatchup'])) * 100
                                                                                                            
                                                                                                            
alldf = alldf[(alldf.Fusion > 0) & (alldf.AquaLS2_Matchup > 0)]

pd.DataFrame(alldf)
```

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import cross_val_predict
diabetes = datasets.load_diabetes()
X = diabetes.data[:150]
y = diabetes.target[:150]
lasso = linear_model.Lasso()

from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

X, y = load_diabetes(return_X_y=True)
lr = LinearRegression()
y_pred = cross_val_predict(lr, X, y, cv=10)
print(y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import PredictionErrorDisplay

fig, axs = plt.subplots(ncols=2, figsize=(8, 4))
PredictionErrorDisplay.from_predictions(
    y,
    y_pred=y_pred,
    kind="actual_vs_predicted",
    subsample=100,
    ax=axs[0],
    random_state=0,
)
axs[0].set_title("Actual vs. Predicted values")
PredictionErrorDisplay.from_predictions(
    y,
    y_pred=y_pred,
    kind="residual_vs_predicted",
    subsample=100,
    ax=axs[1],
    random_state=0,
)
axs[1].set_title("Residuals vs. Predicted Values")
fig.suptitle("Plotting cross-validated predictions")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor

# Load the diabetes dataset
diabetes = load_diabetes()

# Create a random forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the diabetes dataset
rf.fit(diabetes.data, diabetes.target)

# Predict the target values for the first data point
prediction = rf.predict(diabetes.data[:1,:])
print("Prediction:", prediction)

# Get the predictions of all individual trees in the forest
individual_predictions = np.zeros((diabetes.data.shape[0], rf.n_estimators))
for i, tree in enumerate(rf.estimators_):
    individual_predictions[:,i] = tree.predict(diabetes.data)

# Plot the histogram of the predicted values
plt.hist(individual_predictions[0,:])
plt.title("Histogram of Predicted Values for First Data Point")
plt.xlabel("Predicted Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor

# Load the diabetes dataset
diabetes = load_diabetes()
X = diabetes.data
y_true = diabetes.target

# Fit a random forest model with 100 trees
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y_true)

# Predict the first data point
pred = rf.predict(X[0].reshape(1, -1))
tree_preds = rf.estimators_
tree_preds = [tree.predict(X[0].reshape(1, -1))[0] for tree in tree_preds]

# Plot the histograms
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Histogram of predicted values from all trees
ax[0].hist(tree_preds, bins=30, alpha=0.5, label='Tree predictions')
ax[0].axvline(x=pred, color='r', label='Random forest prediction')
ax[0].set_xlabel('Target value')
ax[0].set_ylabel('Frequency')
ax[0].legend()

# Histogram of target variable
ax[1].hist(y_true, bins=30)
ax[1].set_xlabel('Target value')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Histogram of target variable')

plt.show()
```

```python
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor
import numpy as np

# Load the diabetes dataset
diabetes = load_diabetes()

# Define the features and target variable
X = diabetes.data
y = diabetes.target

# Train the random forest regressor model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# Get the predictions of each individual decision tree
tree_predictions = np.stack([tree.predict(X) for tree in rf.estimators_], axis=0)

# Get the prediction of the random forest by averaging the individual tree predictions
rf_prediction = np.mean(tree_predictions, axis=0)

# Plot the histogram of the random forest predictions
import matplotlib.pyplot as plt
plt.hist(rf_prediction, bins=30, alpha=0.5, label='Random Forest')
plt.hist(y, bins=30, alpha=0.5, label='True Values')
plt.legend(loc='upper right')
plt.show()


# Get the histogram values
counts, bin_edges = np.histogram(pred, bins=10)

# Calculate the cumulative sum
cdf = np.cumsum(counts)

# Normalize the cumulative sum
cdf = cdf / cdf[-1]

# Plot the cdf
plt.plot(bin_edges[1:], cdf)

# Set labels and title
plt.xlabel('Predictions')
plt.ylabel('Cumulative Density')
plt.title('Cumulative Density Function of Random Forest Predictions')

plt.show()
```
