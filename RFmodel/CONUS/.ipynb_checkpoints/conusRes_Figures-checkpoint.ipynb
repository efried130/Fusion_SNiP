```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a dictionary of data
data = {
    'Bar 1': {'x': [1, 2, 3, 4, 5], 
              'y': [2, 4, 3, 4, 5], 
              'err': [0.5, 0.8, 0.6, 0.6, 0.7]},
    'Bar 2': {'x': [1, 2, 3, 4, 5], 
              'y': [3, 1, 4, 4, 5], 
              'err': [0.4, 0.6, 0.7, 0.6, 0.7]},
    'Bar 3': {'x': [1, 2, 3, 4, 5], 
              'y': [1, 3, 2, 4, 5], 
              'err': [0.6, 0.9, 0.5, 0.6, 0.7]}
}

# Create the figure and axis objects
fig, ax = plt.subplots()

# Set the bar width
width = 0.25

# Loop over the dictionary keys and values
for i, (label, values) in enumerate(data.items()):
    x = np.array(values['x'])
    y = np.array(values['y'])
    err = np.array(values['err'])

    # Plot the bars
    ax.bar(x + i * width, y, width, label=label)

    # Plot the error bars
    ax.errorbar(x + i * width, y, yerr=err, fmt='none', ecolor='black', capsize=4)

# Add a legend and labels
ax.legend()
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Multiple Bar Chart with Error Bars')

plt.show()
```
