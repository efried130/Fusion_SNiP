```python
#module imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.stats import linregress
import glob
import statsmodels.api as sm


os.chdir('/nas/cee-water/cjgleason/ellie/SNiP/RFmodel/CONUS/')
```

```python
#Sentinel - Landsat Band Comparison

file_list = glob.glob(os.path.join('/nas/cee-water/cjgleason/ellie/SNiP/RFmodel/CONUS/LS2/LS2_data', '*.csv'))
dataframe = pd.concat(map(pd.read_csv, file_list), ignore_index=True)

df=dataframe.dropna(subset='red')

df = df[(df.red > 0) & (df.red < 3000) & (df.nir > 0) & (df.blue > 0) & (df.green > 0) & (df.swir1 > 0) & (df.swir2 > 0) & (df.nir > 0)]

print(df.columns.values.tolist())

# dataframe = pd.read_csv('/nas/cee-water/cjgleason/ellie/SNiP/RFmodel/CONUS/LS2/LS2_data/huc6Sites_2013-03-18_2023-03-04.csv')
# print(dataframe.columns.values.tolist())
# print(dataframe.SiteID.unique())
```

```python
# Define the function to apply to the new 'sat' column
def get_sat(satID):
    if satID.startswith('LT5'):
        return int(5.0)
    elif satID.startswith('LE7'):
        return int(7.0)
    elif satID.startswith('LC8'):
        return int(8.0)
    elif satID.startswith('LC9'):
        return int(9.0)
    elif satID.startswith('L2A'):
        return int(2.0)
    else:
        return 'Other'

# Create a new column 'Region' based on the 'City' column
df['sat'] = df['sceneID'].apply(lambda x: get_sat(x))
print(df.columns.values.tolist())
```

```python
#Avg same day S2
df_A = df[df['sat'] == 2]

avg_df = df_A.groupby(['SiteID', 'date']).mean().reset_index()
#print(avg_df)

# filter groups A and B
df_B = df[df['sat'].isin([5, 7, 8, 9])]

# merge the DataFrames on date
df_merged = pd.merge(avg_df[['date', 'red', 'blue', 'green', 'nir', 
                           'swir1', 'swir2', 'sat', 'SiteID']], df_B[['date', 'red', 'blue', 'green', 'nir', 'swir1', 'swir2',
                                                                                                             'sat', 'SiteID']], on= ['SiteID','date'])
df_merged = df_merged[df_merged['red_x'] < 2500]
df_merged = df_merged[df_merged['red_y'] < 2500]


# output the filtered and merged DataFrame
print(df_merged)

fig, ax = plt.subplots()
p1 = max(max((df_merged.red_x)), max((df_merged.red_y)))
p2 = min(min((df_merged.red_x)), min((df_merged.red_y)))
plt.plot([p1, p2], [p1, p2], 'b-')

# Plot x from group 1 against x from group 2
ax.scatter(df_merged['red_x'], df_merged['red_y'])

# Set x and y axis labels
ax.set_xlabel('Red Band Sentinel')
ax.set_ylabel('Red Band Landsat')

# Set plot title
ax.set_title('Comparison of Red between Landsat Series and Sentinel-2')

# Display the plot
plt.show()

plot = sns.scatterplot(data=df_merged, x = df_merged['red_x'], y = df_merged['red_y'], hue="SiteID", legend=False)
# add 1:1 line
max_value = max(max(df_merged['red_x']), max(df_merged['red_y']))
sns.lineplot(x=[0, max_value], y=[0, max_value], color='black')
#sns.move_legend(plot, "upper left", bbox_to_anchor=(1, 1))

plot
```

```python
df_merged = df_merged[(df_merged.red_x < 2500) & (df_merged.green_x < 2500) & (df_merged.blue_x < 2500) & (df_merged.nir_x < 2500) & (df_merged.swir1_x < 2500) & (df_merged.swir2_x < 2500)]
df_merged = df_merged[(df_merged.red_y < 2500) & (df_merged.green_y < 2500) & (df_merged.blue_y < 2500) & (df_merged.nir_y < 2500) & (df_merged.swir1_y < 2500) & (df_merged.swir2_y < 2500)]

# create some data
x = df_merged['red_x']
y = df_merged['red_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("Red Band Comparison between Landsat Series and Sentinel")
plt.show()


# create some data
x = df_merged['green_x']
y = df_merged['green_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("Green Band Comparison between Landsat Series and Sentinel")
plt.show()

# create some data
x = df_merged['blue_x']
y = df_merged['blue_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("Blue Band Comparison between Landsat Series and Sentinel")
plt.show()

# create some data
x = df_merged['nir_x']
y = df_merged['nir_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("NIR Band Comparison between Landsat Series and Sentinel")
plt.show()

# create some data
x = df_merged['swir2_x']
y = df_merged['swir2_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("Swir1 Band Comparison between Landsat Series and Sentinel")
plt.show()

# create some data
x = df_merged['swir2_x']
y = df_merged['swir2_y']

# fit linear regression model
model = sm.OLS(y, sm.add_constant(x)).fit()

# plot the data and the regression line
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x, model.predict(sm.add_constant(x)), color='red')
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls='--', c='y')
# add the equation and R squared to the plot
ax.text(0.05, 0.95, f"y = {model.params[0]:.2f} + {model.params[1]:.2f}x", transform=ax.transAxes)
ax.text(0.05, 0.9, f"R\u00b2 = {model.rsquared:.2f}", transform=ax.transAxes)

# add axis labels and title
ax.set_xlabel("Sentinel")
ax.set_ylabel("Landsat")
ax.set_title("Swir2 Band Comparison between Landsat Series and Sentinel")
plt.show()
```

```python
from scipy.stats import linregress


x = df_merged['nir_x']
y = df_merged['nir_y']

# Fit a regression line
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Plot the data and the regression line
plt.scatter(x, y)
plt.plot(x, intercept + slope*x, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print('slope:', slope)
print('intercept:', intercept)
print('r_value:', r_value)
print('p_value:', p_value)
print('std_err:', std_err)
```

```python
# Create a new array of corrected y values
x_corrected = ((y - intercept) / slope)

# Plot the data and the regression line (multiplied by two...?)
plt.scatter(x_corrected / 2, y)
plt.plot(x, intercept + slope*x, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import linregress

# create a sample dataframe with 6 columns of x data and 6 columns of y data
df = pd.DataFrame(valMerge_slim, columns=['red_x', 'green_x', 'blue_x', 'nir_x', 'swir1_x', 'swir2_x', 'red_y', 'green_y', 'blue_y', 'nir_y', 'swir1_y', 'swir2_y', 'sat_x'])

# create a figure object and a set of subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
axs = axs.flatten() # flatten the array of subplots to iterate over them

# iterate over the subplots and plot each pair of x and y columns
for i in range(6):
    x_col = df.iloc[:, i] # select the ith column as the x data
    y_col = df.iloc[:, i+6] # select the corresponding y column
    
    # plot the scatterplot and add a line of best fit
    sns.regplot(x=x_col, y=y_col, ax=axs[i], ci=None, scatter_kws={'alpha':0.4, 'color':'purple'}, line_kws={'color':'red', 'label': 'Best Fit Line'})
    axs[i].plot(x_col, x_col, color='blue', linestyle='--', label='1:1 Line') # add 1:1 line
    axs[i].legend(loc = 'lower right') # show legend
    
    # add the equation and R-squared value to the plot
    slope, intercept, r_value, p_value, std_err = linregress(x_col, y_col)
    equation = f'y = {slope:.2f}x + {intercept:.2f}'
    r_squared = f'R^2 = {r_value**2:.2f}'
    axs[i].text(0.05, 0.9, equation, transform=axs[i].transAxes)
    axs[i].text(0.05, 0.8, r_squared, transform=axs[i].transAxes)

# add a suptitle for the entire figure
fig.suptitle('Same-day Fusion vs LS2 Band Reflectances', fontsize=16, fontweight='bold')

# add x and y axis labels
fig.text(0.5, 0.04, 'LS2', ha='center', fontsize=14, fontweight='bold')
fig.text(0.06, 0.5, 'Fusion', va='center', rotation='vertical', fontsize=14, fontweight='bold')

plt.show()
```
