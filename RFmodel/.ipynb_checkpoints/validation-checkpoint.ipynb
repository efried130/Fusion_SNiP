# Validation Images

```python
#module imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import glob

os.getcwd()
```

# POTENTIAL WQP MATCHUPS JUST USING MODIS

```python
file_list = glob.glob(os.path.join('rawMODIS_huc5_all', '*.csv'))
dataframe = pd.concat(map(pd.read_csv, file_list), ignore_index=True)
print(dataframe.columns.values.tolist())
print(len(dataframe.index))
```

```python
dataframe = dataframe.sort_values(by = 'date')
dataframe['date'] = pd.to_datetime(dataframe['date']).dt.date
dataframe['date'] = dataframe.date.apply(lambda x: str(x))
dataframe = dataframe[(dataframe.maskedPercent <= 10) & (dataframe.maskedPercent >= 0)]
print(dataframe.date)
print(len(dataframe.index))
print(dataframe.dtypes)
```

```python
sns.histplot(x = 'SiteID', data = dataframe)

rawSiteCounts = dataframe['SiteID'].value_counts()
print('# of MODIS images per Site:', rawSiteCounts)
```

```python
#Bring in Aquasat HLS
aquaHuc5 = pd.read_csv(r'aquasatHLS_huc5_ratio.csv', low_memory=False)
aquaHuc5.columns = aquaHuc5.columns.str.replace("[.]", "_")
#aquaHuc5['date'] = pd.to_datetime(aquaHuc5['date']).dt.strftime('%Y-%m-%D')
print(aquaHuc5.columns.values.tolist())
aquaHuc5['dateStr'] = aquaHuc5['date']
aquaHuc5['date'] = pd.to_datetime(aquaHuc5['date']).dt.date
print(aquaHuc5.date)
aquaHuc5['Matchup'] = 'AquaSat'
aquaHuc5 = aquaHuc5[['SiteID', 'date', 'tss', 'units', 'Matchup']]
```

```python
#Not aquasat rest of huc5 WQP
huc5 = pd.read_csv(r'notAquasat_huc5_5.csv', low_memory=False)
huc5['tss'] = huc5['value']
huc5['Matchup'] = 'WQP'
huc5 = huc5[['SiteID', 'date', 'tss', 'units', 'Matchup']]

huc5All = pd.concat([aquaHuc5, huc5]).dropna(subset='tss')#.drop_duplicates(subset = ['SiteID', 'date'])

huc5All['date'] = pd.to_datetime(huc5All['date']).dt.date
huc5All['date'] = huc5All.date.apply(lambda x: str(x))
print(huc5All.dtypes)
print(len(huc5All.index))
print(huc5All.date)
```

```python
modisMatchups = pd.merge(dataframe, huc5All, on =['SiteID', 'date'], how ="inner")
print(modisMatchups)
modisMergeSites1 = list(modisMatchups.SiteID.unique())
print(len(modisMergeSites1))
```

```python
#sort the dataframes
huc5All = huc5All.sort_values(by = 'date')
huc5All = huc5All[huc5All['units'] == 'mg/l']
huc5All['date'] = pd.to_datetime(huc5All['date'])

dataframe = dataframe.sort_values(by = 'date')
dataframe['date'] = pd.to_datetime(dataframe['date'])

modisMerge_timediff = pd.merge_asof(dataframe, huc5All, on = 'date', by ='SiteID',
                                 tolerance=pd.Timedelta(1, "d"),
                                 allow_exact_matches=True).dropna(subset = ['tss'])

# print the variable
print(modisMerge_timediff.columns.values.tolist())
print('# of Matchups Across all Sites:', len(modisMerge_timediff.index))

#get siteID's
modisMergeSites = list(modisMerge_timediff.SiteID.unique())
print('# of Unique Sites', len(modisMergeSites))

siteCounts = modisMerge_timediff['SiteID'].value_counts()
print('# of Addtnl MODIS matches per Site:', siteCounts)
```

```python
print(modisMerge_timediff.head())
#modisMerge_timediff.to_csv(r'modisMerge_timediff_TEST2.csv')
```

```python
sns.histplot(x = 'SiteID', data = modisMerge_timediff)
```

```python
huc5_aquasatSites = huc5[huc5['SiteID'].isin(modisMergeSites)].copy()

modisMergedFig = modisMerge_timediff[['SiteID', 'date']]
huc5_aquasatSites['date'] = pd.to_datetime(huc5_aquasatSites['date'])
huc5['date'] = pd.to_datetime(huc5['date'])
aquaHuc5['date'] = pd.to_datetime(aquaHuc5['date'])


huc5_unmatched = pd.merge(modisMergedFig, huc5_aquasatSites, on =['SiteID', 'date'], how ="outer").dropna(subset='tss')
huc5_unmatched = huc5_unmatched[huc5_unmatched['units'] == 'mg/l']
print(huc5_unmatched)

aquaFWqp = pd.concat([huc5_unmatched, aquaHuc5, modisMergedFig]).drop_duplicates(subset = ['SiteID', 'date'])

aquaFWqp = aquaFWqp.groupby(['SiteID', 'Matchup'])['tss'].count().unstack()#.fillna(0)
aquaFWqp.plot(kind='bar', stacked=True, color = ['blue', 'green', 'black'])
plt.ylabel('Measurement Count', fontsize = '20')
plt.xlabel('SiteID', fontsize = '20')
plt.title('Fraction of Available Measurements Matched to Satellite Images')
plt.tick_params(bottom=False, labelbottom=False)
plt.legend(bbox_to_anchor=(1, 1), fontsize='20')
```

# Image List

```python
file_list = glob.glob(os.path.join('validationImgs', '*.csv'))
dataframe = pd.concat(map(pd.read_csv, file_list), ignore_index=True)
print(dataframe.columns.values.tolist())
print(len(dataframe.index))
```

```python
#Drop na, synch column names if needed, and remove negative band values
df=dataframe.dropna(subset='blue').drop_duplicates()
df = df[(df.red > 0) & (df.nir > 0) & (df.blue > 0) & (df.green > 0) & (df.swir1 > 0) & (df.swir2 > 0)]
df['date_time'] = df['date']
df['date'] = pd.to_datetime(df['date']).dt.date
print(len(df.index))
print(df.date)
```

```python
df = df.assign(R_GB = df['red']/ (df['green'] + df['blue']),
          Matchup = 'Validation') 
df = df[['SiteID', 'date', 'R_GB', 'Matchup']]
df.columns.values.tolist()
dfSites = list(df.SiteID.unique())
```

# Merge to AquaFusion

```python
aquaFusion = pd.read_csv(r'r_gb_aquaFusion_16Sites_12_2022.csv')
aquaFusion['date'] = pd.to_datetime(aquaFusion['date']).dt.date
print(aquaFusion.date)
print(len(aquaFusion.index))
aquaFusion = aquaFusion[['SiteID', 'date', 'R_GB', 'Matchup']]

aquaFusion = aquaFusion[aquaFusion['SiteID'].isin(dfSites)]
```

```python
aquaFusVal = pd.concat([aquaFusion, df]).dropna(subset='R_GB')
print(len(aquaFusVal.index))
```

```python
#Generate a first look Plot
plot = sns.scatterplot(data=aquaFusVal, x="date", y="R_GB", style ="Matchup", hue = 'SiteID')
plt.legend(fontsize='6')

sns.move_legend(plot, "upper left", bbox_to_anchor=(1.05, 1))
plot
```
